```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer


class TextModel:
    def __init__(self, string):
        self.string = string

    def get_answer(self, question: str, n_answers: int = 1) -> list[str]:
        cleaned_text = [s.strip() for s in self.string.split('.') if s.strip()]
        vectorizer = TfidfVectorizer()
        tfidf = vectorizer.fit_transform(cleaned_text + [question]).toarray()
        tfidf_text = tfidf[:-1]
        tfidf_question = tfidf[-1:]
        similarities = cosine_similarity(tfidf_text, tfidf_question).flatten()
        sorted_indices = np.argsort(similarities)[::-1]
        sorted_indices = sorted_indices[similarities[sorted_indices] > 0][:n_answers]
        return [cleaned_text[i] for i in sorted_indices]
```

```python
text = ("I want to start working as a programmer. I'm learning programming at Tel-Ran course.\
        The teachers says to us that for becoming a programmer you should perform Homeworks.\
                        But because of family and work I don't have time to do Homework.")

t_model = TextModel(text)
question = "What don't you do Homeworks and What do you want?"
t_model.get_answer(question, 1)
```
