```python
import pandas as pd
df = pd.read_csv('../data/car_sales_cleaned1.csv')
df
```

```python
df = df.drop_duplicates() # removing duplicates
df.shape
```

```python
df = df.dropna() # remove rows with missing values
df.shape
```

```python
valuesColModel = df['Model'].unique()
valuesColModel
```

```python
dfProjection = df[['Annual Income', 'Price ($)']]
dfProjection
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
X = dfProjection[['Annual Income']]
y = dfProjection[['Price ($)']]
model.fit(X.values, y)
prediction = model.predict([[30_000]])
prediction
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model.fit(X_train, y_train)
prediction = model.predict(X_test)
score = accuracy_score(y_test, prediction)
score
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
import converter
model = DecisionTreeClassifier()
df_projection = df[['Company', 'Model']]
mapper = converter.columns_mapper(['Company', 'Model'], df_projection)
df_converted = converter.convert_x(df_projection, mapper)
X = df_converted[['Model']]
y = df_converted[['Company']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model.fit(X_train, y_train)
prediction = model.predict(X_test)
score = accuracy_score(y_test, prediction)
score
```
