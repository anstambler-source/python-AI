```python
import pandas as pd
df = pd.read_csv('../data/car_sales_cleaned1.csv')
df
```

```python
df = df.drop_duplicates() # removing duplicates
df.shape
```

```python
df = df.dropna() # remove rows with missing values
df.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
import joblib
import converter

model = DecisionTreeClassifier()
y = df[['Gender']]
mapper = converter.columns_mapper(['Gender', 'Company', 'Model', 'Transmission', 'Color'], df)
df_converted = converter.convert_x(df, mapper)
X = df_converted[['Model', 'Transmission', 'Color']]
model.fit(X, y)
joblib.dump(model, '../models/car_gender_model.pkl')
```
