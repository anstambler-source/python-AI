```python
import numpy as np
data_set = np.array([
    [0.3, -1.2, 0.5, 0.9],
    [0.7, -2.7, 0.6, 0.1],
    [0.3, 0.01, 0.47, 2],
    [0.5, 2.2, 0.11, 3.0],
    [0.12, -0.7, 0.15, 0.25]
])
vec = np.array([0.6, -1.5, 0.3, 0.2])
vec_len = np.linalg.norm(vec) # vi4islyaem dlinu(normu) vectora
vec_len
```

```python
vec_norm = vec / vec_len # privodim dlinu na6ego vec k 1
vec_norm
```

```python
vec_norm_len = np.linalg.norm(vec_norm)
vec_norm_len
```

```python
# data_set_len = np.linalg.norm(data_set)
# data_set_len
data_set_lens = np.linalg.norm(data_set, axis=1, keepdims=True) # axis 1 - s4itaem dlini strok
data_set_lens
```

```python
data_set_norm = data_set / data_set_lens
data_set_norm
```

```python
data_set_norm_len = np.linalg.norm(data_set_norm, axis=1, keepdims=True)
data_set_norm_len
```

```python
similarities = data_set_norm @ vec_norm
similarities
```

```python
best_index = np.argmax(similarities) # i6em naibolee pohojiy index(1 ideal)
best_index
```
