```python
# 384-d SBERT
# Not our case now
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

text = ("I want to start working as a programmer. I'm learning programming at Tel-Ran course.\
        The teachers says to us that for becoming a programmer you should perform Homeworks.\
                        But because of family and work I don't have time to do Homework.")

text_arr = text.split('.')
cleaned_text = [res for s in text_arr if(res := s.strip())]
cleaned_text
```

```python
# query = ["What do you want?"]
# query = ["Where are you learning programming?"]
# query = ["What does the teachers say to you?"]
query = ["What don't you do Homeworks?"]
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(cleaned_text + query).toarray()
tfidf
```

```python
tfidf_text = tfidf[:-1]
tfidf_query = tfidf[-1:]
similarities = cosine_similarity(tfidf_text, tfidf_query)
best_index = np.argmax(similarities)
cleaned_text[best_index]
```
