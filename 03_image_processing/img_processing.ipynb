```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img1 = np.zeros((100, 100, 3), dtype="uint8")
img1[50][50][0] = 255
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
# doagonal
height = 100
width = 100
img1 = np.zeros((height, width, 3), dtype="uint8")
for i in range(height):
    for j in range(width):
        if i == j:
            img1[i][j][0] = 255
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
# doagonal
height = 100
width = 100
img1 = np.zeros((height, width, 3), dtype="uint8")
cv2.line(img1, (0, 0), (width, height), (255, 0, 0), thickness=1)
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
cv2.line(img1, (0, height), (width, 0), (255, 0, 0), thickness=1)
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
cv2.line(img1, (0, height // 2), (width, height // 2), (255, 0, 0), thickness=1)
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
cv2.circle(img1, (width // 2, height // 2), width // 3, color=(255, 0, 0), thickness=-1)
plt.axis('off')
plt.imshow(img1)
plt.show()
```

```python
img2 = np.random.randint(0, 256, (100, 100, 3), dtype="uint8")
plt.axis('off')
plt.imshow(img2)
plt.show()
```

```python
img3 = np.random.randint(0, 256, (50, 50), dtype="uint8")
plt.axis('off')
plt.imshow(img3, cmap="gray") # color map
plt.show()
```

```python
bg = np.full((100, 100), 200, dtype="uint8")
plt.imshow(bg, cmap="gray", vmin=0, vmax=255) # contrast
plt.show()
```

```python
import random
img4 = np.full((30, 30), 255, dtype="uint8")
for i in range(30):
    for j in range(30):
        img4[i][j] = 0 if random.randint(0, 2) else 200

plt.imshow(img4, cmap="gray", vmin=0, vmax=255)
plt.show()
```
